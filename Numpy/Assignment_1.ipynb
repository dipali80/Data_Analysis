```python
#Q3A) Perform the following operations on an array of mobile phones prices 6999, 7500, 
#11999, 27899, 14999, 9999.
#Create a 1d-array of mobile phones prices 

m_prices= np.array([6999,7500,11999,14999,9999])
m_prices
```

```python
# 3B)  Convert this array to float type 
m_prices.astype(float)
```

```python
# 3C) Append a new mobile having price of 13999 Rs. to this array 
m_prices = np.append(m_prices ,13999)
m_prices
```

```python
#3D)  Reverse this array of mobile phones prices 
m_prices[::-1]

```

```python
#3E) Apply GST of 18% on mobile phones prices and update this array. 
m_price = m_prices*0.18
m_price
```

```python
# 3F) Sort the array in descending order of price 
np.sort(m_price)[::-1]
```

```python
# 3G) What is the average mobile phone price 
np.mean(m_price)
```

```python
# 3H) What is the difference b/w maximum and minimum price 
m_price = np.max(m_price)-np.min(m_price)
m_price
```
