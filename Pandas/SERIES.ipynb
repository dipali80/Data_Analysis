#### performing operation on two series

```python
s1 = pd.Series([10,20,30,40,50])
s2 = pd.Series([1,2,3,4,5])
s1+s2
```

#### performing operation with  labeling ===== (Opertaion perform on only same labeling if the labels are differant then it gives nan values )

```python
s1 = pd.Series([10,20,30,40,50],index=['a','b','c','d','e'])
s2 = pd.Series([1,2,3,4,5],index=['a','b','d','c','z'])
s1+s2 ## it give nan values for z and e bcz it does not match with any one
```
